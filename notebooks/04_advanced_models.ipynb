# 04 - Advanced Models

**Urban Flood Risk Prediction - CSE 572**

This notebook trains advanced models:
1. Random Forest
2. Gradient Boosting
3. XGBoost
4. LightGBM
5. Neural Network

```python
import sys
sys.path.append('..')

from src.data_preprocessing import load_and_preprocess
from src.models import (
    train_random_forest, train_gradient_boosting,
    train_xgboost, train_lightgbm, train_neural_network
)
from src.evaluation import evaluate_model, save_results, print_comparison_table

print("Modules loaded!")
```

```python
# Load data
X_train, X_val, y_train, y_val = load_and_preprocess()
results = {}
```

```python
# Random Forest
(rf_model, rf_time) = train_random_forest(X_train, y_train)
results['Random Forest'] = evaluate_model(rf_model, X_val, y_val, 'Random Forest')
results['Random Forest']['training_time'] = rf_time
```

```python
# XGBoost
(xgb_model, xgb_time) = train_xgboost(X_train, y_train)
results['XGBoost'] = evaluate_model(xgb_model, X_val, y_val, 'XGBoost')
results['XGBoost']['training_time'] = xgb_time
```

```python
# LightGBM
(lgbm_model, lgbm_time) = train_lightgbm(X_train, y_train)
results['LightGBM'] = evaluate_model(lgbm_model, X_val, y_val, 'LightGBM')
results['LightGBM']['training_time'] = lgbm_time
```

```python
# Neural Network
(nn_model, nn_time) = train_neural_network(X_train, y_train, X_val, y_val)
results['Neural Network'] = evaluate_model(nn_model, X_val, y_val, 'Neural Network')
results['Neural Network']['training_time'] = nn_time
```

```python
# Compare all models
print_comparison_table(results)
save_results(results, 'advanced_model_results')
```
