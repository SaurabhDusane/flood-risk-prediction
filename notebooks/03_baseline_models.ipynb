# 03 - Baseline Models

**Urban Flood Risk Prediction - CSE 572**

This notebook trains and evaluates baseline models:
1. SGD Regressor
2. K-Nearest Neighbors Regressor

```python
import sys
sys.path.append('..')

from src.data_preprocessing import load_and_preprocess
from src.models import train_sgd_regressor, train_knn_regressor
from src.evaluation import evaluate_model, print_comparison_table

print("Modules loaded!")
```

```python
# Load preprocessed data
X_train, X_val, y_train, y_val = load_and_preprocess()
```

```python
# Train SGD Regressor
(sgd_model, sgd_time) = train_sgd_regressor(X_train, y_train)
sgd_metrics = evaluate_model(sgd_model, X_val, y_val, 'SGD Regressor')
sgd_metrics['training_time'] = sgd_time
```

```python
# Train KNN Regressor
(knn_model, knn_time) = train_knn_regressor(X_train, y_train)
knn_metrics = evaluate_model(knn_model, X_val, y_val, 'KNN Regressor')
knn_metrics['training_time'] = knn_time
```

```python
# Compare results
baseline_results = {
    'SGD Regressor': sgd_metrics,
    'KNN Regressor': knn_metrics
}
print_comparison_table(baseline_results)
```
